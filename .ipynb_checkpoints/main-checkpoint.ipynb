#### Separating Date and Time

```python
def dt_parser(x, dt='date'):
    if dt=='date':
        return x.split()[0]
    elif dt=='time':
        return x.split()[1]
```

```python
def hour_parser(x):
    return dt_parser(x, 'time').split(':')[0]
```

```python
def date_parser(x):
    return dt_parser(x, 'date').split('-')
```

```python
train.head()
```

```python
date_split = train['datetime'].apply(date_parser).apply(pd.Series)
date_split.columns = ['year', 'month', 'day']
train = pd.concat([train, date_split], axis=1)
train.head()
```

```python
hour_series = train['datetime'].apply(hour_parser)
hour_series.name = 'hour'
train = pd.concat([train, hour_series], axis=1)
```

```python
train.drop('datetime', axis=1, inplace=True)
```

```python
def parse_date_time(df):
    date_split = df['datetime'].apply(date_parser).apply(pd.Series)
    date_split.columns = ['year', 'month', 'day']
    df = pd.concat([df, date_split], axis=1)
    
    hour_series = df['datetime'].apply(hour_parser)
    hour_series.name = 'hour'
    df = pd.concat([df, hour_series], axis=1)
    
    df.drop('datetime', axis=1, inplace=True)
    
    return df
```
