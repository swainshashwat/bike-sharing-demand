#### Separating Date and Time

```python
def dt_parser(x, dt='date'):
    if dt=='date':
        return x.split()[0]
    elif dt=='time':
        return x.split()[1]
```

```python
train['date'], train['time'] = (train['datetime'].apply(dt_parser, dt='date'),
                                train['datetime'].apply(dt_parser, dt='time'))
```
